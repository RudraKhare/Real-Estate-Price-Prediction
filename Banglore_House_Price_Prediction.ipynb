```python
#This function checks if a given input x can be turned into a floating-point number, if yes it will return True and if no it will return False
def is_float(x):
  try:
    float(x)
    return True
  except:
    return False
```

```python
df3[~df3['total_sqft'].apply(is_float)].head()
#This funtion only prints true values so that we put ~. So we can get false values as true. here it is the ranges.
```

```python
def convert_sqft_to_num(x):
  tokens=x.split('-')
  if len(tokens)==2:
    return (float(tokens[0])+float(tokens[1]))/2
  try:
    return float(x)
  except:
    return None
```

```python
# Function that will remove those properties whose prize_per_sqft is either much higher or lower then a typical property prize

def remove_pps_outliers(df):
  df_out=pd.DataFrame()
  for key, subdf in df.groupby('location'):
      m = np.mean(subdf.price_per_sqft)
      st=np.std(subdf.price_per_sqft)
      reduced_df=subdf[(subdf.price_per_sqft>(m-st))&(subdf.price_per_sqft<=(m+st))]
      df_out = pd.concat([df_out,reduced_df],ignore_index=True)
  return df_out

df7=remove_pps_outliers(df6)
```

```python
#This funtion was created to view 2bhk and 3bhk properties in reference with sqft and price per sqft
# This function was created to view 2bhk and 3bhk properties in reference with sqft and price per sqft
def plot_scatter_chart(df, location):
    # Filter DataFrame to get 2 BHK and 3 BHK properties for the specified location
    #These lines create two new DataFrames, bhk2 and bhk3, by filtering the original DataFrame df based on the specified location and the number of bedrooms (bhk).
    #bhk2 contains rows where the location is equal to the specified location and the number of bedrooms is 2.
    #bhk3 contains rows where the location is equal to the specified location and the number of bedrooms is 3.

    bhk2 = df[(df.location == location) & (df.bhk == 2)]
    bhk3 = df[(df.location == location) & (df.bhk == 3)]

    # Set the size of the plot
    #matplotlib.rcParams['figure.figsize'] = (15, 18)

    # Scatter plot for 2 BHK properties
    plt.scatter(bhk2.total_sqft, bhk2.price_per_sqft, color='blue', label='2 BHK', s=50)

    # Scatter plot for 3 BHK properties (corrected line)
    plt.scatter(bhk3.total_sqft, bhk3.price_per_sqft, marker='+', color='green', label='3 BHK', s=50)

    # Set x-axis label
    plt.xlabel("Total Square Feet Area")

    # Set y-axis label
    plt.ylabel("Price Per Square Feet")

    # Set the title of the plot based on the specified location
    plt.title(location)

    # Add a legend to the plot
    plt.legend()

# Call the function with the DataFrame df7 and location "Rajaji Nagar"
plot_scatter_chart(df7, "Rajaji Nagar")

# Display the plot
plt.show()
```

```python
#Since there are properties of 2bhk that have much higher price then 3bhk and considerd to be an outlier.
#Below funtion will do the required
# We can remove those 2bhk apartments whse price_per_sqft is less than price_per_sqft of 1bhk apartment

def remove_bhk_outliers(df):
    exclude_indices = np.array([]) #This line initializes an empty NumPy array named exclude_indices
    for location, location_df in df.groupby('location'): # The variable location represents each unique 'location', and location_df represents the DataFrame containing data only for that specific 'location'.
        bhk_stats = {} #This line initializes an empty dictionary named bhk_stats
        for bhk, bhk_df in location_df.groupby('bhk'): #The variable bhk represents each unique 'bhk' (number of bedrooms), and bhk_df represents the DataFrame containing data only for that specific 'bhk' within the current 'location' group.

            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
            }#Creates a dictionary entry for the current 'bhk' in the bhk_stats dictionary, storing mean, standard deviation, and count.

        for bhk, bhk_df in location_df.groupby('bhk'): #This line starts another loop over 'bhk' groups within the 'location' group.
            if bhk > 1:    # Checks if the current 'bhk' is greater than 1.
                stats = bhk_stats.get(bhk - 1) #Retrieves statistics for the previous 'bhk' group (if it exists) from the bhk_stats dictionary.
                if stats and stats['count'] > 5: # Checks if there are statistics for the previous 'bhk' group and the count is greater than 5.whether there are at least 6 data points in the previous 'bhk' group.It ensures that there is a sufficiently large sample size for statistical measures to be meaningful.
                    exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft < (stats['mean'])].index.values) #rows with 'price_per_sqft' values below the mean of the previous 'bhk' group are considered outliers.

    return df.drop(exclude_indices, axis='index')
```

We will check the efficiency of model with different Estimators to chosse which one works best.

```python
from sklearn.model_selection import GridSearchCV, ShuffleSplit #GridSearchCV is used for hyperparameter tuning, ShuffleSplit is a cross-validator for shuffling and splitting the data
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.tree import DecisionTreeRegressor
import pandas as pd

def find_best_model_using_gridsearchcv(X, y):
    algos = {
        'linear_regression': {
            'model': LinearRegression(),
               'params':{}
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1, 2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion': ['mse', 'friedman_mse'],
                'splitter': ['best', 'random']
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
      #algo_name: This is a variable name that represents the current key (algorithm name) in each iteration of the loop. It takes on the values of the keys in the algos dictionary.
      #config: This is another variable name that represents the current value (configuration dictionary) associated with the current key in each iteration. It takes on the values of the corresponding values (dictionaries) in the algos dictionary.

        gs = GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        #config['model'] is used to access the model associated with the current algorithm.
        #config['params'] is used to access the hyperparameters associated with the current algorithm.
        gs.fit(X,y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })
    return pd.DataFrame(scores, columns=['model', 'best_score', 'best_params'])

find_best_model_using_gridsearchcv(X,y)
```
